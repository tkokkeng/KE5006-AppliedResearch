```python
%matplotlib inline
import matplotlib.pyplot as plt
import os
import pickle
```

## 1x16cv-2x16gruB-1x8d, dropout 060606, warm restart cycle = 10

```python
with open(os.path.join('models', 'pcp-1x16cv-2x16gruB-1x8d-dropout060606-wr1-histories-epch1200.pkl'), 'rb') as infile:
    histories = pickle.load(infile)
```

```python
train_loss = []
val_loss = []
train_acc = []
val_acc = []
for h in histories:
    for a_loss in h['loss']:
        train_loss.append(a_loss)
    for a_loss in h['val_loss']:
        val_loss.append(a_loss)
    for a_acc in h['acc']:
        train_acc.append(a_acc)
    for a_acc in h['val_acc']:
        val_acc.append(a_acc)
epochs = range(1, len(train_loss) + 1)
plt.figure(figsize=(16, 6))
plt.subplot(121)
plt.plot(epochs, train_loss, 'g', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.title('1x16 Conv1D 2x16 GRU Bidirectional 1x8 Dense\nwith Dropouts (.6/.6/.6), warm restarts cycle=10\nTraining and Validation Loss')
plt.legend()
plt.subplot(122)
plt.plot(epochs, train_acc, 'g', label='Training Accuracy')
plt.plot(epochs, val_acc, 'b', label='Validation Accuracy')
plt.title('1x16 Conv1D 2x16 GRU Bidirectional 1x8 Dense\nwith Dropouts (.6/.6/.6), warm restarts cycle=10\nTraining and Validation Accuracy')
plt.legend()
plt.show()
```

```python
start, stop = 500, 550
epochs = range(start, stop)
plt.figure(figsize=(16, 6))
plt.subplot(121)
plt.plot(epochs, train_loss[slice(start, stop)], 'g', label='Training Loss')
plt.plot(epochs, val_loss[slice(start, stop)], 'b', label='Validation Loss')
plt.title('1x16 Conv1D 2x16 GRU Bidirectional 1x8 Dense\nwith Dropouts (.6/.6/.6), warm restarts cycle=10\nTraining and Validation Loss')
plt.legend()
plt.subplot(122)
plt.plot(epochs, train_acc[slice(start, stop)], 'g', label='Training Accuracy')
plt.plot(epochs, val_acc[slice(start, stop)], 'b', label='Validation Accuracy')
plt.title('1x16 Conv1D 2x16 GRU Bidirectional 1x8 Dense\nwith Dropouts (.6/.6/.6), warm restarts cycle=10\nTraining and Validation Accuracy')
plt.legend()
plt.show()
```
